```python
import requests
import string
import csv
import re
from collections import Counter

# Download the text file from Project Gutenberg
url = "https://www.gutenberg.org/cache/epub/5827/pg5827.txt"
response = requests.get(url)
text = response.text

# Tokenization and word processing
translator = str.maketrans('', '', string.punctuation)
words = text.split()  # Don't convert to lowercase here
processed_words = [word.translate(translator) for word in words]

# Count word frequencies
word_counts = Counter(processed_words)

# Create all_words.csv with 'utf-8-sig' encoding
with open("all_words.csv", "w", newline="", encoding="utf-8-sig") as csvfile:
    csvwriter = csv.writer(csvfile)
    csvwriter.writerow(["Word", "Frequency"])
    for word, count in sorted(word_counts.items()):
        csvwriter.writerow([word, count])



# Define a regular expression pattern to match capitalized words
pattern = r'\b[A-Z][A-Za-z\']*'

# Find all capitalized words in the text
uppercase_words = re.findall(pattern, text)
processed_words = [word.translate(translator) for word in uppercase_words]

# Count word frequencies
word_counts = Counter(processed_words)

# Create all_uppercase_words.csv with 'utf-8-sig' encoding
with open("all_uppercase_words.csv", "w", newline="", encoding="utf-8-sig") as csvfile:
    csvwriter = csv.writer(csvfile)
    csvwriter.writerow(["Word", "Frequency"])
    for word, count in sorted(word_counts.items()):
        csvwriter.writerow([word, count])

# To find Which were the 5 most frequent words in all_words.csv were most frequent?


# Load the CSV file containing all words
with open("all_words.csv", "r", newline="", encoding="utf-8-sig") as csvfile:
    csvreader = csv.reader(csvfile)
    next(csvreader)  # Skip the header row
    word_counts = [(word, int(count)) for word, count in csvreader]

# Sort words by frequency (in descending order)
sorted_word_counts = sorted(word_counts, key=lambda x: x[1], reverse=True)

# Get the 5 most frequent words
top_5_words = sorted_word_counts[:5]

# Print the 5 most frequent words
print("5 most frequent words in all_words.csv are as follows")
for word, count in top_5_words:
    print(f"{word}: {count}")

# To find Which were the 5 most frequent words in all_uppercase_words.csv were most frequent?

# Load the CSV file containing all uppercase words
with open("all_uppercase_words.csv", "r", newline="", encoding="utf-8-sig") as csvfile:
    csvreader = csv.reader(csvfile)
    next(csvreader)  # Skip the header row
    uppercase_word_counts = [(word, int(count)) for word, count in csvreader]

# Sort uppercase words by frequency (in descending order)
sorted_uppercase_word_counts = sorted(uppercase_word_counts, key=lambda x: x[1], reverse=True)

# Get the 5 most frequent uppercase words
top_5_uppercase_words = sorted_uppercase_word_counts[:5]

# Print the 5 most frequent uppercase words
print("5 most frequent words in all_uppercase_words.csv are as follows")
for word, count in top_5_uppercase_words:
    print(f"{word}: {count}")

```
